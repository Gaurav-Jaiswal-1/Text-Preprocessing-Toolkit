```python
import sys, os
import pandas as pd
from sklearn.datasets import fetch_20newsgroups

# Ensure src/ is in path (important when running from experiments/test.ipynb)
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from src.TextPreprocessingToolkit.text_preprocessor import TextPreprocessor

data = fetch_20newsgroups(subset='train', categories=['rec.autos'], remove=('headers', 'footers', 'quotes'))

# Create DataFrame and limit to 1000 samples
df = pd.DataFrame({'review': data.data[:1000], 'label': data.target[:1000]})

# Save raw data
raw_path = r"E:\Text-Preprocessing-Toolkit\data\text_data\news_raw.csv"
df.to_csv(raw_path, index=False)

tp = TextPreprocessor(spell_correction=False)
tp.process_csv(
    input_path=raw_path,
    text_column="review",
    output_path=r"E:\Text-Preprocessing-Toolkit\cleaned_data\cleaned_text_data\news_clean.csv",
    steps=['clean', 'punctuation', 'lowercase', 'tokenize', 'stopwords', 'lemmatize']
)

clean_df = pd.read_csv(r"E:\Text-Preprocessing-Toolkit\cleaned_data\cleaned_text_data\news_clean.csv")

print("Before:\n", df['review'].iloc[0][:200])
print("\nAfter:\n", clean_df['review'].iloc[0])
```
